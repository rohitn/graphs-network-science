```python
import pandas as pd
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/tomasonjo/graphs-network-science/main/dataset/hetionet/edges.csv")
```

```python
data.head()
```

```python
data['type'].unique()
```

```python
from pykeen.triples import TriplesFactory

tf = TriplesFactory.from_labeled_triples(
  data[["source_name", "type", "target_name"]].values,
)
```

```python
training, testing, validation = tf.split([.8, .1, .1])
```

```python
from pykeen.pipeline import pipeline

result = pipeline(
    training=training,
    testing=testing,
    validation=validation,
    model='PairRE',
    stopper='early',
    epochs=100,
    dimensions=128
)
```

```python
result.plot_losses()
```

```python
result.get_metric("AMR")
```

```python
result.get_metric("AMRR")
```

```python
result.get_metric("ahits@k")
```

```python
from pykeen.models.predict import get_all_prediction_df

top_k_predictions_df = get_all_prediction_df(result.model, k=15000, triples_factory=result.training, remove_known=True)
print(top_k_predictions_df[top_k_predictions_df['relation_label'] == 'treats'].head(5))
```
