```python
import pandas as pd

data = pd.read_csv(
    "https://bit.ly/3X2qp1r"
)
data.head()
```

```python
data["type"].unique()
```

```python
from pykeen.triples import TriplesFactory

tf = TriplesFactory.from_labeled_triples(
    data[["source_name", "type", "target_name"]].values,
)
```

```python
training, testing, validation = tf.split([0.8, 0.1, 0.1], random_state=42)
```

```python
from pykeen.pipeline import pipeline

result = pipeline(
    training=training,
    testing=testing,
    validation=validation,
    model="PairRE",
    stopper="early",
    epochs=100,
    random_seed=42,
)
```

```python
from pykeen.models.predict import get_tail_prediction_df

pred_df = get_tail_prediction_df(
    result.model,
    head_label="Acetylsalicylic acid",
    relation_label="treats",
    remove_known=True,
    triples_factory=result.training,
)
pred_df.head()
```
